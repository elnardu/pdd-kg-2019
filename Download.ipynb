```python
import json
import requests
```

```python
with open('cookies.json') as f:
    cookies = json.load(f)
```

```python
s = requests.Session()
for cookie in cookies:
    s.cookies.set(name=cookie['name'], value=cookie['value'])
    
HEADERS = {
    'User-Agent': ''
}
```

```python
def download_item(item_id):
    res = s.get(f'https://demoprava.srs.kg/dst/play/play.html',
            params={'itemId': item_id}, headers=HEADERS)
    
    print(f'GET {item_id} {res.status_code}')
    
    if res.status_code != 200:
        return
    
    with open(f'docs/{item_id}.html', 'w') as f:
        f.write(res.text)
        
```

```python
for i in range(8000):
    download_item(i)
```
