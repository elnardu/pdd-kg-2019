```python
correctValueRe = re.compile(r"item\.readAttribute\('value'\) == '([^']+)'")

def create_card(file_path, skip_media_download=False):
    note_id = os.path.splitext(os.path.basename(file_path))[0]

    
    with open(file_path) as f:
        text = f.read()
    
    soup = BeautifulSoup(text, 'html.parser')
    
    res = soup.find(id='choiceInteraction-RESPONSE')
    
    images = res.find_all('img')
    for i, image in enumerate(images):
        src = image['src']
        if not skip_media_download:
            fetch_asset(src, f'{note_id}_{i}')
        image['src'] = f'{note_id}_{i}.png'
    
    
    before = str(res)
    correctValue = correctValueRe.search(text).group(1)
    soup.find(value=correctValue).parent.parent['class'].append('correct-choice')
    after = str(res)
    return before, after
    
create_card('docs/178.html');
```

```python
my_model = genanki.Model(
    1202994046,
    'Simple Model',
    fields=[
        {'name': 'Question'},
        {'name': 'Answer'},
    ],
    templates=[
        {
            'name': 'Card 1',
            'qfmt': '{{Question}}',
            'afmt': '{{Answer}}',
        },
    ],
    css='.correct-choice {color: green; font-weight: bold;}')
```

```python
os.makedirs('anki', exist_ok=True)
```

```python
shutil.rmtree('assets')
os.makedirs('assets', exist_ok=True)

my_deck = genanki.Deck(
    1865573765,
    'ПДД КР 2019 (Русский)')


for i in range(5000):
    path = PATH_TO_DOCS + f'/{i}.html'
    if not os.path.exists(path):
        continue

    res = create_card(path)
        
    my_note = genanki.Note(
        model=my_model,
        fields=res)
    
    my_deck.add_note(my_note)
    
    
my_package = genanki.Package(my_deck)
my_package.media_files = [
    'assets/' + f for f in os.listdir('assets')
]
my_package.write_to_file('anki/traffic_code_kg_russian_2019.apkg')
```
